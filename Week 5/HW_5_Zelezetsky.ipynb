```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cvx
np.random.seed(5)
# number of stocks
n = 500
# number of factors (not described or needed in problem statement)
k = 20

# generate rbar, Sigma
# factor matrix, entries uniform on [-0.5,1]
F = 1.5 * np.random.rand(n, k) - 0.5
Sigma = 0.5 * F @ np.diag(np.random.rand(k)) @ F.T + np.diag(0.1 * np.random.rand(n))
# risk-free return (weekly, 5% annual)
mu_rf = 0.2
# Sharpe ratio
SR = 0.4
# expected return
rbar = mu_rf + SR * np.sqrt(np.diag(Sigma))
# market capitalization (index weights)
c = 5 + np.exp(2 * np.random.randn(n))
```

##### векор r можно оценить через вектор его средних(Мю)

    Теперь решим это через CVXPY c добавлением L1 регуляризации , так как именно она облдает тем свойством,
    которе нам необходимо : она более агрессивно(по сравнению с L2 Reg) заталкивает веса в ноль

```python
def f(w):
    return cvx.norm1(w)

w = cvx.Variable(n, nonneg=True)
srr = Sigma+np.outer(rbar,rbar)

constraints = [cvx.quad_form(c-w,srr) - 0.01*cvx.quad_form(c,srr) <= 0]

obj = cvx.Minimize(f(w))
prob = cvx.Problem(obj,constraints)
prob.solve()


print('Кол-во итераций: {}'.format(prob.solution.attr['num_iters']))
```

```python
print(f"Число элементов в решении, по модулю больших 0.01: {sum(abs(w.value)>=0.01)}")
print(f"Число элементов в в-ре c, по модулю больших 0.01: {sum(abs(c)>=0.01)}")
```

    а также его ковариации с другими активами, ведь если они большие, то по одному активу можно делать 
    устойчивые предположения о других, ведь они сильно зависимы.

Сформулированная ниже задача является выпуклой, так как целевая функция ее имеет следующий вид:

$$
\frac{(c-w)^T(\Sigma + \mu\mu^T) (c-w)}{c^T(\Sigma + \mu\mu^T)c}+\lambda\lVert w \rVert_1
$$
Снизу дроби константа,а в числителе квадратичная форма(вып.функция).В сочетании с первой нормой даёт вып.функцию(корень из исходной постановки я отбросил так как он не на что не влияет, кроме того считая градиент я буду отбрасывать знаменатель так как он тоже никакой роли не игает в силу того что является константой)

#### Перейдём к методу проксимального градиента

```python
A = (Sigma + np.outer(rbar,rbar))

B = (cvx.quad_form(c , A)).value 

def f(x):
    return ((cvx.quad_form(c-x , A)).value /B)+lam*cvx.norm1(x).value

def grad(x):
    return ((A.T + A) @ (c - x))/B


def prox_g(x):
    return np.sign(x) * np.maximum(abs(x) - lam*alpha, 0)
```

```python
def pgm(grad, prox_g, x0, alpha, num_iter):
    conv = [x0.copy()]
    x = x0.copy()
    for i in range(num_iter):
        x = prox_g(x - alpha * grad(x))
        conv.append(x.copy())
    return x, conv

def fpgm(grad, prox_g, x0, alpha, num_iter):
    conv = [x0.copy()]
    x_prev = x0.copy()
    y = x0.copy()
    for i in range(num_iter):
        x_next = prox_g(y - alpha * grad(y))
        y = x_next + (i+1) / (i+4) * (x_next - x_prev)
        x_prev = x_next.copy()
        conv.append(x_next.copy())
    return abs(x_next), conv
```

```python
np.random.seed(55)
x0 = np.random.uniform(0,1,n)
```
