Исходная задача:
$$ \min\limits_{x \in Q} \frac{1}{2} x^T x + c^Tx$$
$$ s.t. Ax \preccurlyeq b $$
$$x \in \mathbb R^n $$
$$b \in \mathbb R^m $$
$$A \in \mathbb R^{m\times n} $$
Заменим ограничение на барьерную функцию и поместим её в целевую функцию
$$ F(x) = -\sum_{i=1}^m log(-(Ax-b)_{i})
= -\sum_{i=1}^m log((b-Ax)_{i})
$$
Тогда задача примет следующий вид:
$$ \min\limits_{x \in Q} \frac{1}{2} x^T x + c^Tx -\sum_{i=1}^m log((b-Ax)_{i})$$
$$x \in \mathbb R^n $$
$$b \in \mathbb R^m $$
$$A \in \mathbb R^{m\times n} $$
Градиент барьера:
$$\nabla F(x) =  \sum_{i=1}^m \frac{a_{i}}{ b_{i}-a_{i}^Tx }
$$
Гессиан барьера:
$$\nabla^2 F(x) = \sum_{i=1}^m \frac{a_{i}a_{i}^T}{(b_{i}-a_{i}^Tx)^2}
$$
Кроме того, градиент и гессиан можно переписать в более компактном виде для удобства дальнейших преобразований. Скажем, что :
$$\nabla F(x) = \sum_{i=1}^m \frac{a_{i}}{ b_{i}-a_{i}^T } = A^Td
$$
$$\nabla^2 F(x) = \sum_{i=1}^m \frac{a_{i}a_{i}^T}{(b_{i}-a_{i}^Tx)^2} = A^T diag(d)^2 A
$$
где $d_{i} = \frac{1}{b_i -a_i^T x}$

Теперь мы готовы перенести всё в целевую функцию и взять её градиент и гессиан, которые будем использовать в дальнейшем для решения системы равнений :
$$\nabla f(x) = t(x + c) + A^T d $$
$$\nabla^2 f(x) = tI + A^T diag(d)^2 A$$

Шаг метода Ньютона в общем виде: $$ x_{k+1} = x_{k}-[\nabla^2f(x)]^{-1}\nabla f(x)  $$
Тогда : $$ x_{k+1}-x_{k}= -[\nabla^2f(x)]^{-1}\nabla f(x)  $$
$$ \Delta x= -[\nabla^2f(x)]^{-1}\nabla f(x)  $$
$$ \nabla^2f(x)\Delta x= -\nabla f(x)  $$
В нашей задаче это принимает вид :
$$ (tI + A^T diag(d)^2 A)\Delta x= -(t(c) + A^T d)  $$

##### 1) При m>>n:
Подсчёт гессиана $(tI + A^T diag(d)^2 A)$ подсчитывается следующим образом: так как матрица $diag(d)^2$ является диагональной и представляет из себя $\sum_{i=1}^m \frac{a_{i}a_{i}^T}{(b_{i}-a_{i}^Tx)^2}$, то тогда размерность её равна $m\times m$.
Следовательно разметность 
$ \underbrace{A^T}_{n\times m}  \underbrace{diag(d)^2}_{m \times m}\underbrace{A}_{m \times n}= n\times n$ Сложность вычисления оценивается в $O(mn^2)$
Далее поскольку матрица $(tI + A^T diag(d)^2 A)$ является положительно определённой, то уместно применить разложение Холецкого для решения данной системы.Доказанная сложность решения этой системы $O(\frac{1}{3}n^3)$

##### Анализ сложности одной итерации:
    Для определения общей сложности операции нам надо
        сравнить О(mn^2) и O(1/3n^3). 
        И тут появляются два варианта:
        1)Кол-во ограничений m фиксировано и растёт только размерность 
            задачи n
        2)m строго связана с n какой-то пропорцией и с ростом n растёт и m
Если вариант 1 является верным, то ассимптотически $O(\frac{1}{3}n^3)$ растёт быстрее и следовательно сложность итерации оценить нужно ей. Если вариант 2 является верным, то однозначного ответа тут нет, так как нам неизвестна пропорция между m и n. Может так оказаться, что m настолько больше n, что даже куб в $O(\frac{1}{3}n^3)$ не позволит ей выйти вперёд в плане роста.
##### Поэтому оценим нашу сложность грубо как $mn^2 + \frac{1}{3}n^3$

##### 2) При m << n:
Сложность кубическая, однако можно ее понизить :
$$ \left [ \begin{matrix} tI & A^T \\  A& -diag(d)^{-2} \end{matrix}\right]  \left [ \begin{matrix} \Delta x\\  \nu \end{matrix}\right] = \left [ \begin{matrix} -tc - A^T d\\  0 \end{matrix}\right]$$
$$ (diag(d)^{-2} + \frac{1}{t}A A^T) \nu = -Ac - \frac{1}{t}A A^T d 
$$
По прежнму не зная условий соотношения m и n мы не можем назвать точную стоимость, так как не знаем, подвижна ли m.Тогда так же грубо оценим сложность операций как $nm^2 + \frac{1}{3}m^3$
